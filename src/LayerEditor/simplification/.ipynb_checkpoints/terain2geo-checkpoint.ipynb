```python
import numpy as np
import csv
from scipy.interpolate import griddata
import math
# Import ZlomovÃ½ch dat:
def read_csv_points(inFilename, sort=1, sort2=0):
    with open(inFilename+'.csv', 'r') as csvfile:
        data=csv.reader(csvfile, delimiter=',')
        header=next(data, None)
        header = [x for x in header if x]
        points=np.zeros_like(header)
        for row in data:
            points=np.vstack((points,row))
        points=points[1:,:]
        if sort:
            points=points[points[:, 2].argsort(),:]
        if sort2:
            out=np.zeros_like(points[0,:])
            for order in np.unique(points[:,2]):
                out=np.vstack((out,points[points[:,2]==order,:]))
            out=np.delete(out, 0, 0)
            points=out
    return header, points

def write_polygon_geo(outFilename,points):
    with open(outFilename+'.geo','w') as outfile:
        num=0
        for point in points:
            num+=1
            outfile.write(r'xph_'+str(num)+' = '+str(point[0])+';  ')
            outfile.write(r'yph_'+str(num)+' = '+str(point[1])+';  ')
            outfile.write('  \n')
        outfile.write('pocet_bodu_hranice = '+str(num)+'; \n')
    print(('Wrote '+str(num)+' points in file: '+str(outFilename)+'.geo'))
    return False

def write_polygons_geo(outFilename,points):
    polId=np.arange(np.unique(points[:,2]).shape[0])+1
    nbPoiPol=0
    with open(outFilename+'.geo','w') as outfile:
        for num,point in enumerate(points):
            nbPoiPol += 1
            outfile.write(r'xl_'+str(polId[point[2]==np.unique(points[:,2])][0])+'_'+str(nbPoiPol)+' = '+str(point[0])+';  ')
            outfile.write(r'yl_'+str(polId[point[2]==np.unique(points[:,2])][0])+'_'+str(nbPoiPol)+' = '+str(point[1])+';  ')
            outfile.write('  \n')
            if np.any(point!=points[-1]):
                if (point[2]!=points[num+1,2]):
                    outfile.write('nbPointsLit_'+str(polId[point[2]==np.unique(points[:,2])][0])+' = '+str(nbPoiPol)+'; \n')
                    nbPoiPol-=num+1
            else:
                outfile.write('nbPointsLit_'+str(polId[point[2]==np.unique(points[:,2])][0])+' = '+str(nbPoiPol)+'; \n')
    print(('Wrote '+str(num)+' points in file: '+str(outFilename)+'.geo'))
    return False

def write_tectonics_geo(outFilename,points,names=['t','nbClefts']):
    tmpx=[]
    tmpy=[]
    num=0
    tecOrder=1
    cleftNb=0   
    out=""
    done=True
    for idx,point in enumerate(points):
        num+=1
        tmpx.append(point[0])
        tmpy.append(point[1])
        if np.any(point!=points[-1]):
            if point[3]!=points[idx+1,3]:
                print("Posledni bod pukliny")
                cleftNb+=1
#                 if done and cleftNb==33:
#                     done = False
#                     cleftNb-=1
#                     continue
                out+=r'x'+names[0]+'_'+str(tecOrder)+'_'+str(cleftNb)+' = {'+', '.join([str(x) for x in tmpx])+'}; '
                out+=r'y'+names[0]+'_'+str(tecOrder)+'_'+str(cleftNb)+' = {'+', '.join([str(x) for x in tmpy])+'}; '
                out+='\n'
                tmpx=[]
                tmpy=[]
            if point[2]!=points[idx+1,2]:
                print(("Posledni puklina typu: "+str(tecOrder)))
                tmpx=[]
                tmpy=[]
                out+=names[1]+'_'+str(tecOrder)+' = '+str(cleftNb)+'; \n'
                cleftNb=0
                tecOrder+=1
        else:
            #Last point
            cleftNb+=1
            print("Posledni bod pukliny")
            out+=r'x'+names[0]+'_'+str(tecOrder)+'_'+str(cleftNb)+' = {'+', '.join([str(x) for x in tmpx])+'}; '
            out+=r'y'+names[0]+'_'+str(tecOrder)+'_'+str(cleftNb)+' = {'+', '.join([str(x) for x in tmpy])+'}; '
            out+='\n'
            print(("Posledni puklina typu: "+str(tecOrder)))
            out+=names[1]+'_'+str(tecOrder)+' = '+str(cleftNb)+'; \n'
    with open(outFilename+'.geo','w+') as outfile:
        outfile.write(out)
    print(('Wrote '+str(num)+' points in file: '+str(outFilename)+'.geo'))
    return False

def write_lithology_geo(outFilename,points,litNames=['l','nbLithTypes']):
    tmpx=[]
    tmpy=[]
    nbLithTypes=0
    out=""
    print((points[:,0:4]))
    for idx,point in enumerate(points):
        tmpx.append(point[0])
        tmpy.append(point[1])
        if np.any(point!=points[-1]):
            if point[2]!=points[idx+1,2]:
                nbLithTypes+=1
                out+=r'x'+litNames[0]+'_'+point[2]+' = {'+', '.join([str(x) for x in tmpx])+'}; '
                out+=r'y'+litNames[0]+'_'+point[2]+' = {'+', '.join([str(x) for x in tmpy])+'}; '
                out+='\n'
                tmpx=[]
                tmpy=[]
        else:
            nbLithTypes+=1
            out+=r'x'+litNames[0]+'_'+point[2]+' = {'+', '.join([str(x) for x in tmpx])+'};'
            out+=r'y'+litNames[0]+'_'+point[2]+' = {'+', '.join([str(x) for x in tmpy])+'};'
            out+='\n'
    out+=litNames[1]+' = '+str(nbLithTypes)+'; \n'
    with open(outFilename+'.geo','w') as outfile:
        outfile.write(out)
    print(('Wrote lithology points in file: '+str(outFilename)+'.geo'))
    return False

def checkDuplicates(points):
    uniPoints,uniIdx=np.unique(points[:,0:2], return_counts=True, axis=0)
    if not uniPoints.shape==points[:,0:2].shape:
        delIdx=[]
        for i,dupl in enumerate(uniPoints[uniIdx!=1,:]):
            tmp=np.where(np.all(points[:,0:2]==dupl,axis=1))[0]
            delIdx.insert(i,tmp[1])
        print(('Following list of duplicates found (indeces of duplicates only): ', delIdx))
        points=np.delete(points, tuple(i for i in delIdx), axis=0)
        return points
        
    else:
        print('All good, no duplicates found.')
        return points

def processTerrain(inFilename,outFilename,extend=[1e3,1e3,1e3,1e3],step=100):
    p = np.fromfile('Input/'+inFilename,sep="\t")
    p = p.reshape(-1,3)
    mask = p[:,2]>1e5
    p[mask,2] = np.interp(np.flatnonzero(mask), np.flatnonzero(~mask), p[~mask,2])

    grid_x, grid_y = np.mgrid[p[:,0].min()-extend[0]:p[:,0].max()+extend[1]:step, p[:,1].min()-extend[2]:p[:,1].max()+extend[3]:step]
    grid_z0 = griddata(p[:,0:2], p[:,2], (grid_x, grid_y), method='nearest')
    
    np.savetxt('Output/'+outFilename,np.hstack((grid_x.reshape(-1,1), grid_y.reshape(-1,1),grid_z0.reshape(-1,1))),delimiter='\t')
    return False
```

# Hranice oblasti

```python
#possible boundary solution

import fiona
import pylab as pl
import shapely.geometry as geometry
from descartes import PolygonPatch
from shapely.ops import cascaded_union, polygonize
from scipy.spatial import Delaunay
import numpy as np
import math
from rdp import rdp

def alpha_shape(points, alpha):
    """
    Compute the alpha shape (concave hull) of a set
    of points.
    @param points: Iterable container of points.
    @param alpha: alpha value to influence the
        gooeyness of the border. Smaller numbers
        don't fall inward as much as larger numbers.p
        Too large, and you lose everything!
    """
    if len(points) < 4:
        # When you have a triangle, there is no sense
        # in computing an alpha shape.
        return geometry.MultiPoint(list(points)).convex_hull
    def add_edge(edges, edge_points, coords, i, j):
        """
        Add a line between the i-th and j-th points,
        if not in the list already
        """
        if (i, j) in edges or (j, i) in edges:
            # already added
            return
        edges.add( (i, j) )
        edge_points.append(coords[ [i, j] ])
    coords = np.array([point.coords[0] for point in points])
    alpha=alpha/(max(coords[:,0])-min(coords[:,0])+max(coords[:,1])-min(coords[:,1]))*2
    tri = Delaunay(coords)
    edges = set()
    edge_points = []
    # loop over triangles:
    # ia, ib, ic = indices of corner points of the
    # triangle
    for ia, ib, ic in tri.vertices:
        pa = coords[ia]
        pb = coords[ib]
        pc = coords[ic]
        # Lengths of sides of triangle
        a = math.sqrt((pa[0]-pb[0])**2 + (pa[1]-pb[1])**2)
        b = math.sqrt((pb[0]-pc[0])**2 + (pb[1]-pc[1])**2)
        c = math.sqrt((pc[0]-pa[0])**2 + (pc[1]-pa[1])**2)
        # Semiperimeter of triangle
        s = (a + b + c)/2.0
        # Area of triangle by Heron's formula
        area = math.sqrt(s*(s-a)*(s-b)*(s-c))
        circum_r = a*b*c/(4.0*area)
        # Here's the radius filter.
        #print circum_r
#         print(circum_r)
        if circum_r < 1.0/alpha:
            add_edge(edges, edge_points, coords, ia, ib)
            add_edge(edges, edge_points, coords, ib, ic)
            add_edge(edges, edge_points, coords, ic, ia)
    m = geometry.MultiLineString(edge_points)
    triangles = list(polygonize(m))
    return cascaded_union(triangles)

def plot_polygon(polygon):
    fig = pl.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    margin = .3
    x_min, y_min, x_max, y_max = polygon.bounds
    ax.set_xlim([x_min-margin, x_max+margin])
    ax.set_ylim([y_min-margin, y_max+margin])
    patch = PolygonPatch(polygon, fc='#999999',
                         ec='#000000', fill=True,
                         zorder=-1)
    ax.add_patch(patch)
    return fig

def oneCall(p,alpha):
    global concave_hull
    concave_hull = alpha_shape(p,alpha)
    plot_polygon(concave_hull)
    x = [point.coords.xy[0] for point in p]
    y = [point.coords.xy[1] for point in p]
    pl.plot(x,y,'x', color='#f16824')
    pl.show()
    return False

def straightenBoundary(input_shapefile):
    shapefile = fiona.open(input_shapefile)
    points = [geometry.shape(point['geometry'])
          for point in shapefile]
    print("Input alpha value 'n' to break the loop");
    while True:
        alpha=eval(input("(float) alpha = "))
        if (alpha=="n"):
            break
        try:
            alpha=float(alpha)
        except:
            print("Unable to convert input as float value!")
            break
        oneCall(points,alpha)
    print("Concave hull output straightened boundary:")
    arr = np.array(concave_hull.boundary.xy)
    pl.plot(arr[0,:],arr[1,:])
    pl.show()
    return arr.T

def simplifyBoundary(arr):
    print("Input alpha value 'n' to break the loop");
    while True:
        eps=eval(input("(float) epsilon: "))
        if (eps=="n"):
            break
        try:
            eps=float(eps)
        except:
            print("Unable to convert input as float value!")
            break
        new=rdp(arr,epsilon=eps)
        print(("Number of boundary points: ",new.shape[0]))
        pl.plot(new[:,0],new[:,1])
        pl.show()
    print("Final output simplified boundary:")
    pl.plot(new[:,0],new[:,1])
    pl.show()
    return new
```

```python
input_shapefile = 'Input/Oblast_points.shp'
output_geo = 'Output/Polygon_points'
array = straightenBoundary(input_shapefile)
new = simplifyBoundary(array)
new=checkDuplicates(new)
write_polygon_geo(output_geo,new)
```
